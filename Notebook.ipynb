## Popularity based Recommendations

```python
ratings_merged_books = ratings.merge(books , on= 'ISBN')
```

```python
num_rating_df = ratings_merged_books.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns = {'Book-Rating':'Num-of-Ratings'},inplace = True)
num_rating_df
```

```python
avg_rating_df = ratings_merged_books.groupby('Book-Title').mean(numeric_only = True)['Book-Rating'].reset_index() 
avg_rating_df.rename(columns={'Book-Rating':'avg_rating'},inplace=True)
avg_rating_df
```

```python
popular_df = num_rating_df.merge(avg_rating_df , on = 'Book-Title')
popular_df
```

```python
popular_df = popular_df[popular_df['Num-of-Ratings']>=250].sort_values('avg_rating' , ascending = False).head(50)
```

```python
popular_df = popular_df.merge(books , on = 'Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','Num-of-Ratings','avg_rating']]
#drop_dup as same book name have multiple isbn 
```

```python
popular_df.shape
```

## Collaborative Filtering Based Recommender

```python
# users with greater than 200 ratings
x = ratings_merged_books.groupby('User-ID').count()['Book-Rating']>200
knowledgable_users = x[x].index
```

```python
filtered_rating = ratings_merged_books[ratings_merged_books['User-ID'].isin(knowledgable_users)]
```

```python
# ratings with atleast 50 ratings 
y = filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books = y[y].index
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
pt = final_ratings.pivot_table(index = 'Book-Title' , columns = 'User-ID' , values = 'Book-Rating')
```

```python
pt.fillna(0 , inplace = True)
```

```python
pt
```

### Cosine Simimarity 

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt)
```

```python
def recommend(book_name):
    index = np.where(pt.index == book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])) , key = lambda x:x[1] , reverse = True)[1:5]

    data = []
    for i in similar_items:
        item = []
        temp_df = books[books['Book-Title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))

        data.append(item) # for jinja 

    return data
```

```python
recommend('1984')
```
